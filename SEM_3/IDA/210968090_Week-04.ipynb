## 4) For each of the numeric attributes (proteins up to vitamins), identify and replace all noisy data with the median of attribute.  

```python
def remove_outlier_IQR(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    df_final=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
    return df_final
```

```python
df[[ 'calories', 'protein', 'fat', 'sodium', 'fiber',
       'carbo', 'sugars', 'potass', 'vitamins']]=remove_outlier_IQR(df[[ 'calories', 'protein', 'fat', 'sodium', 'fiber',
       'carbo', 'sugars', 'potass', 'vitamins']])
```

```python
for x in [ 'calories', 'protein', 'fat', 'sodium', 'fiber',
       'carbo', 'sugars', 'potass', 'vitamins']:
    df[x] = df[x].replace(np.nan,df[x].median())
 
```
